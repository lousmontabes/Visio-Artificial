### 1.1 Image creation, saving and loading

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).


a) 
- Create and display a grayscale image that has 150 rows and 200 columns where each value should be of type unsigned integer with 8 bits. 

```python
from skimage import color

# Create a 150*200 matrix composed of a horizontal gradient ranging from values 0 - 200 
gradient_matrix = np.array([[n for n in range(0, 200)] for i in range(0, 150)], dtype = np.uint8)

# Create a 150*200 matrix composed of ones
ones_matrix = np.ones((150, 200), dtype = np.uint8)

# Show a visual representation of both matrices
plt.imshow(gradient_matrix, cmap = "gray")
plt.show()

plt.imshow(ones_matrix, cmap = "gray")
plt.show()
```

**Solution:** 

This allows us to represent 2^8 = 256 values in each pixel.

b) Change the color of the pixels of the image in such a way that the left half is black (0) and the right half is white (255). Visualize the image and print the values of its first row. 

<img src="images_notebook/solution_11b.png">

```python
# Set the second vertical half of the image to white
ones_matrix[0:, int(ones_matrix.shape[1] / 2):] = 255

# Show result
plt.imshow(ones_matrix, cmap = "gray")
plt.show()
```

c) Use the IO image package to save the image previously created:

```python
from skimage import io

# Save image
io.imsave("black_white.jpg", ones_matrix)
```

b) Load the saved image and visualize it:

```python
# Load image
loaded_matrix = io.imread("black_white.jpg")

# Show loaded image
plt.imshow(loaded_matrix, cmap = "gray")
plt.show()
```

### 1.3 Images manipulation

a) Add a black frame to Lenna's image. It should have a width of 1/10th of the height of the original image. 

(We recommend to create a bigger black image to later embed the original image in the central part)

<img src="images_notebook/lena_frame.png">

```python
import math
margin = math.ceil(lenna.shape[0] * .1)
canvas = np.zeros((lenna.shape[0] + 2 * margin, lenna.shape[1] + 2 * margin, 3), dtype = np.uint8)

lenna_framed = canvas.copy()
lenna_framed[margin:canvas.shape[0] - margin, margin:canvas.shape[1] - margin] = lenna

plt.imshow(lenna_framed)
plt.show()
```

b) 
- Change the mask of the framed image to red. 
- Visualize the three images with their respective titles.

```python
canvas[:,:,0] = 255

lenna_framed_red = canvas.copy()
lenna_framed_red[margin:canvas.shape[0] - margin, margin:canvas.shape[1] - margin] = lenna

plt.imshow(lenna_framed_red)
plt.show()

f, axarr = plt.subplots(ncols=3, nrows=1,figsize=(10,5))

axarr[0].imshow(lenna)
axarr[0].set_title('Original')

axarr[1].imshow(lenna_framed, cmap = "gray")
axarr[1].set_title('Black frame')

axarr[2].imshow(lenna_framed_red, cmap = "gray")
axarr[2].set_title('Red frame')

plt.show()
```

### 1.5 Color manipulation

a) 

- Create and visualize the three images shown in the following figure:
<img src="./images_notebook/ex17a.png">

- Using these three images, construct the following color image:
<img src='./images_notebook/ex17b.png'>

```python
# Generate 200x200 left black, right white
vertical_matrix = np.ones((200, 200, 3), dtype = np.uint8)
vertical_matrix[0:, int(vertical_matrix.shape[1] / 2):] = 255
plt.imshow(vertical_matrix, cmap="gray")

# Generate 200x200 top black, bottom white
horizontal_matrix = np.ones((200, 200, 3), dtype = np.uint8)
horizontal_matrix[int(horizontal_matrix.shape[0] / 2):, 0:] = 255
plt.imshow(horizontal_matrix, cmap="gray")

# Generate 200x200 first quarter white, rest black matrix
quarter_matrix = np.ones((200, 200, 3), dtype = np.uint8)
quarter_matrix[:int(quarter_matrix.shape[0] / 2), :int(quarter_matrix.shape[1] / 2)] = 255
plt.imshow(quarter_matrix, cmap="gray")

# Paint white section of vertical matrix red
red_matrix = vertical_matrix.copy()
red_matrix[:,:,1:] = 0
plt.imshow(red_matrix, cmap="gray")

# Paint white section of horizontal matrix green
green_matrix = horizontal_matrix.copy()
green_matrix[:,:,0::2] = 0
plt.imshow(green_matrix, cmap="gray")

# Paint white section of quarter matrix blue
blue_matrix = quarter_matrix.copy()
blue_matrix[:,:,:2] = 0
plt.imshow(blue_matrix, cmap="gray")

# Add up colored matrices
final_matrix = red_matrix + green_matrix + blue_matrix
plt.imshow(final_matrix, cmap="gray")
plt.show()
```

c) 

- Given the images `coat.png` and `model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). 

- Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is only an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying resize, rotate, etc. keep track how the image type and values range change. When necessary, apply type conversion using img_as_float, img_as_uint, img_as_ubyte.

```python
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.color import rgba2rgb

coat = io.imread("images/coat.png")
model = io.imread("images/model.png")
texture = io.imread("images/texture.jpg")

# Resize custom to fit our image
texture = resize(texture, (model.shape[0], model.shape[1]))

# Make a copy of base image
model_coat = model.copy()

# Replace pixels in model_coat where coat isn't black with the equivalent pixels in coat
model_coat[coat != 0] = coat[coat != 0]

# Using the same method, give our model's coat a fresh new fancy look
model_tex = rgba2rgb(model_coat).copy()
coat = rgba2rgb(coat)

model_tex[coat != 0] = texture[coat != 0]

# Display results
f, axarr = plt.subplots(ncols=4, nrows=1,figsize=(15,30))

axarr[0].imshow(coat, cmap = "gray")
axarr[0].set_title('Coat')

axarr[1].imshow(model, cmap = "gray")
axarr[1].set_title('Model')

axarr[2].imshow(model_coat, cmap = "gray")
axarr[2].set_title('Model with coat')

axarr[3].imshow(model_tex, cmap = "gray")
axarr[3].set_title('Model with stylish coat')

plt.show()
```
